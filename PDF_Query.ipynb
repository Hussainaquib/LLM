```python
# Install required packages (if not installed already)
# pip install langchain faiss-cpu transformers pypdf sentence-transformers

from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA

from transformers import pipeline

# 1. Load PDF
loader = PyPDFLoader("/content/drive/MyDrive/1706.03762v7.pdf")
documents = loader.load()

# 2. Split into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)

# 3. Create embeddings (local, no API)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# 4. Store in FAISS vector DB
vectorstore = FAISS.from_documents(docs, embeddings)

# 5. Define local mini LLM (small model, runs locally)
generator = pipeline("text2text-generation", model="google/flan-t5-base")


llm = HuggingFacePipeline(pipeline=generator)

# 6. Build RetrievalQA chain
qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever(),
    chain_type="stuff"
)

# 7. Ask questions about the PDF
while True:
    query = input("\nEnter your question about the PDF (or type 'exit' to quit): ")
    if query.lower() in ["exit", "quit", "q"]:
        print("Goodbye! 👋")
        break

    answer = qa.run(query)
    print(f"\nQ: {query}\nA: {answer}")


print("Q:", query)
print("A:", answer)
```
